```python
from collections import defaultdict

def solution(logs):
    user = []
    problems = defaultdict()
    answer = []
    
    for log in logs:
        user_name, problem = log.split()
        
        if user_name not in user:
            user.append(user_name)
        
        if problem in problems:
            if user_name not in problems[problem]:
                problems[problem].append(user_name)
        else:
            problems[problem] = [user_name]
    
    for i in problems.items():
        if len(i[1]) >= (len(user) / 2):
            answer.append(i[0])
    
    return sorted(answer)     
        
        
        
```

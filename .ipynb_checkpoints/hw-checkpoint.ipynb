## 문제
코딩 1타 강사 민재는 학교에서 코딩을 가르치고 있다. 중간고사때 아이들의 성적을 기록하는게 매우 귀찮았어서, 민재는 기말고사에는 프로그램을 만들어 간편하게 학생들의 성적을 관리해보려 한다!

학생들의 성적을 관리해주는 프로그램을 만들어보자!

## 기능
해당 프로그램은 학생들의 점수를 받으면, 점수와 함께 이를 A,B,C,D,F의 Grade로 변환하여 저장한다. 성적 환산 기준은 90점 이상이면 A, 90점 미만 80점 이상이면 B, 80점 미만 70점 이상이면 C, 70점 미만 60점 이상은 D, 60점 미만은 모두 F이다.

또, 성적을 조회 하면 모든 학생의 이름과 점수, Grade가 이름순으로 내림차순 정렬되어 총인원수와 함께 표시된다.

점수 변경 기능을 통해서 점수 변경도 가능하다. 만약, 학생의 성적이 변경되었다면 이에따라 Grade도 자동으로 해당 점수에 맞는 Grade로 바뀌어서 저장된다. 이와 같이 기능하는 프로그램을 코딩해보자.

* 총인원을 구할때는 len() 메서드를 사용한다. len() 메서드에 대해 공부해본다.

* 이름순으로 정렬할때는 sort() 메서드를 사용한다. sort() 메서드에 대해 공부해본다. 추가로 key를 이용하여 sort()를 사용하는 방법도 익혀보자. 

* 모든 기능을 하나의 함수로 구현하려 하지말고, 각각의 기능을 다른 함수로 구분하고 이 함수들을 main이라는 함수하나에서 작동하게 하면 훨씬 가독성도 좋고, 코딩이 쉬워진다.





````python
profile=dict()
import operator
global c

````

````python
def add():
    a=input('학생의 이름과 점수를 "학생이름,점수"와 같은 형식으로 입력해주세요. 0을 입력받으면 해당 기능이 종료됩니다.').split(',')
    if a[0]=='0':
        return
    if a[0] in profile:
        return('중복된 이름입니다. 다시 입력하세요')
    profile[a[0]]=int(a[1])
    if profile[a[0]]>=90:
        profile[a[0]]=[profile[a[0]],'A']
    elif profile[a[0]]>=80:
        profile[a[0]]=[profile[a[0]],'B']
    elif profile[a[0]]>=70:
        profile[a[0]]=[profile[a[0]],'C']
    elif profile[a[0]]>=60:
        profile[a[0]]=[profile[a[0]],'D']
    elif profile[a[0]]<60:
        profile[a[0]]=[profile[a[0]],'F']


    
    
    
    
````

````python
def view():
    nprofile=sorted(profile.items(), key=operator.itemgetter(1),reverse=True)
    print('이름 점수 Grade')
    for i in nprofile:
        print(i)
    print('총인원은%s명입니다.'%len(nprofile))
        

````

````python
def revise():
    a=input('학생의 이름과 점수를 "학생이름,점수"와 같은 형식으로 입력해주세요. 0을 입력받으면 해당 기능이 종료됩니다.').split(',')
    if a=='0':
        return
    profile[a[0]]=int(a[1])
    if profile[a[0]]>=90:
        profile[a[0]]=[profile[a[0]],'A']
    elif profile[a[0]]>=80:
        profile[a[0]]=[profile[a[0]],'B']
    elif profile[a[0]]>=70:
        profile[a[0]]=[profile[a[0]],'C']
    elif profile[a[0]]>=60:
        profile[a[0]]=[profile[a[0]],'D']
    elif profile[a[0]]<60:
        profile[a[0]]=[profile[a[0]],'F']

    
````

### 피드백 코드

````python
import operator
# import를 제외한 부분은 아예 없는게 좋음!
# 전역변수는 정말 정말 특별한 경우를 제외하고는 사용안함
# 전역변수를 남발하게 되면 변수이름이 겹치게 되고, 하나의 변수를 모든 곳에서 사용할 수록 코드가 변수하나에 종속되므로 유지보수도 복잡해짐
# 최대한 분리하고 또 분리해서 문제가 생기면 문제가 생긴곳만 고칠 수 있도록 항상 기능과 변수를 분리해야함

## 따라서 이부분은 마지막 classManagement 부분으로 이동하고 각자 함수들이 매개변수로 받는게 맞음

def classManagement():
    # 함수나 변수 이름은 항상 소문자와 대문자로 단어를 분리해주어야함 
    # 위와 같이 classManagement와 같은 방식을 "카멜 표기법"이라고 하는데 보통 자바에서 이 변수명 표기를 사용한다.
    # 파이썬에서는 주로 class_management와 같은 "스네이크 표기법"을 사용함 나는 요즘 자바가 손에 익어서 그냥 카멜표기법으로 하겠음
    profile = dict()
    
    while 1:
        selectNum = input("1. 성적 추가\n2. 성적 조회\n3. 성적 변경\n4. 프로그램 종료\n")
        # 변수이름 c로 대충하지말고 이름 제대로 지어라 뒤지기 싫으면 ㅡㅡ
        # 역슬래시 써버릇해라 뒤지기 싫으면 ㅡㅡ 
        if selectNum == '1':
            add(profile)
            
        elif selectNum == '2':    
            view(profile)
            
        elif selectNum == '3':
            revise(profile)
            
        elif selectNum == '4':
            return('프로그램이 종료됩니다.')
        
        else:
            print('다시 입력하세요.')
        
classmanagement()
````

### 피드백 코드

````python
def view(profile):
    sortedProfile = sorted(profile.items())
    # 이 부분은 어차피 이름을 기준으로 정렬하는거라 key 필요없음 key는 맨 앞 요소가 아닌 다른 요소를 기준으로 정렬할 때 필요함
    print('이름\t점수\tGrade')
    
    for i in sortedProfile:
        print(i[0],"\t",i[1][0], "\t", i[1][1])
        
    print('총 인원은 %s명입니다.\n' %len(profile))
````

````python
import operator
# import를 제외한 부분은 아예 없는게 좋음!
# 전역변수는 정말 정말 특별한 경우를 제외하고는 사용안함
# 전역변수를 남발하게 되면 변수이름이 겹치게 되고, 하나의 변수를 모든 곳에서 사용할 수록 코드가 변수하나에 종속되므로 유지보수도 복잡해짐
# 최대한 분리하고 또 분리해서 문제가 생기면 문제가 생긴곳만 고칠 수 있도록 항상 기능과 변수를 분리해야함

## 따라서 이부분은 마지막 classManagement 부분으로 이동하고 각자 함수들이 매개변수로 받는게 맞음

def add(profile):
    while 1:
        # 문제보면 0 입력할 때 까지는 계속 학생 정보 받는건데 if studentInfo[0] == '0' 만들어 놓기는 한거 보니까 까먹은듯??
        print("학생의 이름과 점수를 ""학생이름,점수""와 같은 형식으로 입력해주세요. 0을 입력받으면 해당 기능이 종료됩니다.\n")
        # 되도록이면 print랑 input을 분리해서 쓰는게 좋음, 그리고 ''는 한 문자에만 쓰고, 문장에는 ""을 사용해야함 
        # 파이썬은 개꿀 언어라 ''와 ""의 구분이 없지만 다른 언어에서는 엄연하게 char과 str로 구분됨
        studentInfo = input().split(',')
        # 변수이름 a같이 하지말고 항상 의미 부여 해주기
        
        if studentInfo[0]=='0':
            return

        if studentInfo[0] in profile:
            return("중복된 이름입니다. 다시 입력하세요\n")
        # 여기도 마찬가지로 문장이므로 큰따옴표("") 사용

        profile[studentInfo[0]] = int(studentInfo[1])
        # 연산자 사이에 띄어쓰기 습관 들여야함 가독성에서 차이 정말 많이남

        if profile[studentInfo[0]] >= 90:
            profile[studentInfo[0]] = [profile[studentInfo[0]],'A']

        elif profile[studentInfo[0]] >= 80:
            profile[studentInfo[0]] = [profile[studentInfo[0]],'B']

        elif profile[studentInfo[0]] >= 70:
            profile[studentInfo[0]] = [profile[studentInfo[0]],'C']

        elif profile[studentInfo[0]] >= 60:
            profile[studentInfo[0]] = [profile[studentInfo[0]],'D']

        elif profile[studentInfo[0]] < 60:
            profile[studentInfo[0]] = [profile[studentInfo[0]],'F']
        # 여기도 if 문사이에 항상 띄어서 써야 가독성이 좋음


def view(profile):
    sortedProfile = sorted(profile.items())
    # 이 부분은 어차피 이름을 기준으로 정렬하는거라 key 필요없음 key는 맨 앞 요소가 아닌 다른 요소를 기준으로 정렬할 때 필요함
    print('이름\t점수\tGrade')
    
    for i in sortedProfile:
        print(i[0],"\t",i[1][0], "\t", i[1][1])
        
    print('총 인원은 %s명입니다.' %len(profile))
    
def revise(profile):
    print("학생의 이름과 점수를 ""학생이름,점수""와 같은 형식으로 입력해주세요. 0을 입력받으면 해당 기능이 종료됩니다.\n")
    studentInfo = input().split(',')
    
    if studentInfo == '0':
        return
    
    profile[studentInfo[0]] = int(studentInfo[1])
    
    if profile[studentInfo[0]] >= 90:
        profile[studentInfo[0]] = [profile[studentInfo[0]],'A']
        
    elif profile[studentInfo[0]] >= 80:
        profile[studentInfo[0]] = [profile[studentInfo[0]],'B']
        
    elif profile[studentInfo[0]] >= 70:
        profile[studentInfo[0]] = [profile[studentInfo[0]],'C']
        
    elif profile[studentInfo[0]] >= 60:
        profile[studentInfo[0]] = [profile[studentInfo[0]],'D']
        
    elif profile[studentInfo[0]] < 60:
        profile[studentInfo[0]] = [profile[studentInfo[0]],'F']

def classManagement():
    # 함수나 변수 이름은 항상 소문자와 대문자로 단어를 분리해주어야함 
    # 위와 같이 classManagement와 같은 방식을 "카멜 표기법"이라고 하는데 보통 자바에서 이 변수명 표기를 사용한다.
    # 파이썬에서는 주로 class_management와 같은 "스네이크 표기법"을 사용함 나는 요즘 자바가 손에 익어서 그냥 카멜표기법으로 하겠음
    profile = dict()
    
    while 1:
        selectNum = input("1. 성적 추가\n2. 성적 조회\n3. 성적 변경\n4. 프로그램 종료\n")
        # 변수이름 c로 대충하지말고 이름 제대로 지어라 뒤지기 싫으면 ㅡㅡ
        # 역슬래시 써버릇해라 뒤지기 싫으면 ㅡㅡ 
        if selectNum == '1':
            add(profile)
            
        elif selectNum == '2':    
            view(profile)
            
        elif selectNum == '3':
            revise(profile)
            
        elif selectNum == '4':
            return('프로그램이 종료됩니다.')
        
        else:
            print('다시 입력하세요.\n')
        
classManagement()
````
